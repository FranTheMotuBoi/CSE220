```python
class Patient:
    def __init__(self, id, name, age, blood_group, next, prev):
        self.id = id
        self.name = name
        self.age = age
        self.bg = blood_group
        self.next = next
        self.prev = prev
class WRM:
    def __init__(self):
        self.head = Patient(None, None, None, None, None, None)
        self.head.next = self.head
        self.head.prev = self.head
        self.t = self.head

    def RegisterPatient(self, id, name, age, bg):
        n_p = Patient(id, name, age, bg, self.head, self.t)
        self.t.next = n_p
        self.head.prev = n_p
        self.t = n_p
        print("Successfully Register Patient")

    def ServePatient(self):
        if self.head.next != self.head:
            print(f"{self.head.next.name}, ID: {self.head.next.id}.")
        else:
            print("No patient left.")

        r_n = self.head.next
        if r_n.next == self.head:
            self.head.next = self.head
            self.head.prev = self.head
            self.t = self.head
        else:
            n_node = r_n.next
            self.head.next = n_node
            n_node.prev = self.head
            r_n.next = None
            r_n.prev = None

    def CancelAll(self):
        self.head.next = self.head
        self.head.prev = self.head
        self.t= self.head
        print("Cancelled all appointment.")

    def CanDoctorGoHome(self):
        result="Yes" if self.head.next == self.head else "No"
        print(result)

    def ShowAllPatient(self):
        t = self.head.next
        while t != self.head:
            if t.next != self.head:
                print(t.id, end = "-->")
            else:
                print(t.id)
            t = t.next

# Driver Code print(f'1. Register Patient\n2. Serve Patient\n3. Cancel All\n4. Can Doctor Go Home?\n5. Show details of all patient\n6. exit\n=========================')

appointment_management=WRM()
l=0
while l<2:
  print("WRM")
  print("1. Register Patient")
  print("2. Serve Patient")
  print("3. Cancel All")
  print("4. Can Doctor Go Home?")
  print("5. Show details of all patient")
  print("6. exit")
  user_data=int(input('Please enter your choice: '))
  if user_data==1:
    id=input('Enter an ID: ')
    name=input('Enter a name: ')
    age=int(input('Enter the age: '))
    blood_group=input('Enter the Blood Group: ')
    appointment_management.RegisterPatient(id, name, age, blood_group)
    print()
  elif user_data==2:
    appointment_management.ServePatient()
    print()
  elif user_data==3:
    print('All appoinments are canceled')
    appointment_management.CancelAll()
    print()
  elif user_data==4:
    appointment_management.CanDoctorGoHome()

  elif user_data==5:
    appointment_management.ShowAllPatient()
  elif user_data==6:
    print('EXIT...')
    break
  else:
    print('Please enter a valid option')
```
